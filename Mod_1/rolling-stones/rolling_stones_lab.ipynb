```python
import csv 
from collections import OrderedDict

with open('data.csv') as f:
        reader = csv.DictReader(f)
        albums_info = []
        for row in reader:
            albums_info.append(row)
            
        #data = [collections.OrderedDict(r) for r in reader]
        #data = list(reader)
            
print(albums_info[6])
```

```python
def find_by_name(album_name):
    for index in range(len(albums_info)):
        if albums_info[index]['album'] == album_name:
            return(dict.values(albums_info[index]))
            
    else:
          return None 
```

```python
find_by_name('Revolver')
```

```python
def find_by_rank(album_rank):
    for index in range(0, len(albums_info)):
        if albums_info[index]['number'] == album_rank:
            return(dict.values(albums_info[index]))
            
    else:
          return None 
```

```python
find_by_rank('500')
```

```python
def find_by_year(album_year):
    albums_in_year = []
    empty_list = []
    for index in range(0, len(albums_info)):
        if albums_info[index]['year'] == album_year:
            albums_in_year.append(albums_info[index]['album'])
    
        else: 
            None 
        
    return albums_in_year
```

```python
find_by_year('1999')
```

```python
def find_by_years(start_year, end_year):
    albums_in_year = []
    for index in range(0, len(albums_info)):
        if int(start_year) <= int(albums_info[index]['year']) <= int(end_year):
            albums_in_year.append(albums_info[index]['album']) 
            
        else:
            None 

    return albums_in_year
```

```python
find_by_years(1999, 2000)
```

```python
find_by_years('1997', '2000')
```

```python
def find_by_ranks(start_rank, end_rank):
    albums_in_ranks = []
    for index in range(0, len(albums_info)):
        if int(start_rank) <= int(albums_info[index]['number']) <= int(end_rank):
            albums_in_ranks.append(albums_info[index]['album']) 
            
        else:
            None 

    return albums_in_ranks
```

```python
find_by_ranks('55', '56')
```

```python
def all_titles():
    album_titles = []
    for index in range(0,len(albums_info)):
        album_titles.append(albums_info[index]['album'])
        
    return album_titles
```

```python
def all_artists():
    album_artists = []
    for index in range(0,len(albums_info)):
        album_artists.append(albums_info[index]['artist'])
        
    return album_artists
```

```python
#Artists with the most albums 
#Returns the artist with the highest amount of albums on the list of top albums

def artists_most_albums():
    artists_album_count = {}
    for index in range(0, len(all_artists())):
        if all_artists()[index] in artists_album_count:
            artists_album_count[all_artists()[index]] += 1
        else:
            artists_album_count[all_artists()[index]] = 1
         
    return artists_album_count
    max_albums= max(artists_album_count.values())
    
    max_albums_artists = []
    for k,v in artists_album_count.items():
        if v == max_albums:
            max_albums_artists.append(k)
    return max_albums_artists
```
